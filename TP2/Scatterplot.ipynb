```python
from save_and_load import *
import matplotlib.pyplot as plt
from geo.calculate_distance import calculate_distance_between_ips
```

```python
def add_distances(path):
    ret = []
    for (ttl1, ttl2, ip1, ip2, rtt) in path:
        distance = calculate_distance_between_ips(ip1, ip2)
        ret.append((ttl1, ttl2, ip1, ip2, rtt, distance))
        print(ret[-1])
    return ret
```

```python
from math import log
paths = []
files = [
    "rtt_between_jumps_cam.ac.uk.txt",
    "rtt_between_jumps_jhu.edu.txt",
    "rtt_between_jumps_uj.ac.za.txt",
    "rtt_between_jumps_www.u-tokyo.ac.jp.txt"
]
for f in files:
    paths += add_distances(read_object_from_file(f))
xs = [p[4] for p in paths]
ys = [p[5] for p in paths]
plt.annotate("  [1]", (xs[-2], ys[-2]))
plt.scatter(xs,ys,marker='o')
plt.xlabel('RTT (ms)')
plt.ylabel('Distance (km)')
plt.title("Análisis de correlación distancia-tiempo")
plt.savefig("scatter.png")
plt.show()
```
